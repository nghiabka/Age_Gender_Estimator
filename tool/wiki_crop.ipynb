```python
instances = mat['wiki'][0][0][0].shape[1]
 
columns = ["dob", "photo_taken", "full_path", "gender", "name", "face_location", "face_score", "second_face_score"]
 
import pandas as pd
df = pd.DataFrame(index = range(0,instances), columns = columns)

for i in mat:
    if i == "wiki":
        current_array = mat[i][0][0]
        for j in range(len(current_array)):
            df[columns[j]] = pd.DataFrame(current_array[j][0])
```

```python
df
```

```python
from datetime import datetime, timedelta
def datenum_to_datetime(datenum):
    days = datenum % 1
    hours = days % 1 * 24
    minutes = hours % 1 * 60
    seconds = minutes % 1 * 60
    exact_date = datetime.fromordinal(int(datenum)) \
    + timedelta(days=int(days)) + timedelta(hours=int(hours)) \
    + timedelta(minutes=int(minutes)) + timedelta(seconds=round(seconds)) \
    - timedelta(days=366)

    return exact_date.year
 
df['date_of_birth'] = df['dob'].apply(datenum_to_datetime)
```

```python
df
```

```python
df['age'] = df['photo_taken'] - df['date_of_birth']
```

```python
df
```

```python
#remove pictures does not include face
df = df[df['face_score'] != -np.inf]
 
#some pictures include more than one face, remove them
df = df[df['second_face_score'].isna()]
 
#check threshold
df = df[df['face_score'] >= 3]
 
#some records do not have a gender information
df = df[~df['gender'].isna()]
 
df = df.drop(columns = ['name','face_score','second_face_score','date_of_birth','face_location'])
```

```python
#some guys seem to be greater than 100. some of these are paintings. remove these old guys
df = df[df['age'] <= 100]
 
#some guys seem to be unborn in the data set
df = df[df['age'] > 0]
```

```python
dict_data = {}
data = []
for i in range(df.shape[0]):
    line = df.iloc[i]
    path = line['full_path'][0]
    gender = int(line['gender'])
    age = line['age']
    dict_data['path'] = path
    dict_data['age'] = str(age)
    dict_data['gender'] = str(gender)
    data.append(dict_data)
    dict_data = {}
```

```python
def fix_age_label(age):
    age = int(age)
    try:
        if 0 < age < 13:
            return str(0)

        elif 12 < age < 19:
            return str(1)

        elif 18 < age < 23:
            return str(2)

        elif 22 < age < 30:
            return str(3)

        elif 29 < age < 35:
            return str(4)

        elif 34 < age < 40:
            return str(5)

        elif 39 < age < 45:
            return str(6)
        elif 44 < age < 51:
            return str(7)

        elif 50 < age < 60:
            return str(8)

        else:
            return str(9)

    except Exception as e:
        raise Exception(" {}".format(e))
```

```python
prefix_path = '../wiki_crop/'
new_folder_path = '../wiki_crop/wiki_data/'
for image in tqdm(data):
    path = image['path']
    full_path = os.path.join(prefix_path, path)
    gender = image['gender']
    age = fix_age_label(image['age'])
    age_path = os.path.join(new_folder_path, age)
    gender_path = os.path.join(age_path, gender)
    if not os.path.exists(age_path):
        os.mkdir(age_path)
    if not os.path.exists(gender_path):
        os.mkdir(gender_path)
    copy(full_path, gender_path)
    
```
